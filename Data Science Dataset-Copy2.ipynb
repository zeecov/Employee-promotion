```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import classification_report, accuracy_score 
```

```python
predictions=predictions.astype(int)
```

```python
print(classification_report(val_y, predictions))
```

```python
print(accuracy_score(val_y, predictions))
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_true=val_y, y_pred=predictions)
print('Confusion matrix:\n', conf_mat)

labels = ['Class 0', 'Class 1']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(conf_mat, cmap=plt.cm.Blues)
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Expected')
plt.show()
```

```python
count_class_0, count_class_1 = data.Promoted_or_Not.value_counts()
```

```python
df_class_0 = data[data["Promoted_or_Not"] ==0]
df_class_1 = data[data["Promoted_or_Not"] ==1]
```

```python
df_class_0_under = df_class_0.sample(count_class_1)
df_test_under = pd.concat([df_class_0_under, df_class_1], axis=0)

print('Random under-sampling:')
print(df_test_under.Promoted_or_Not.value_counts())

df_test_under.Promoted_or_Not.value_counts().plot(kind='bar', title='Count (target)')
```

```python
###carrying out oversampling on the dataset
df_class_1_over = df_class_1.sample(count_class_0, replace = True)
df_test_over = pd.concat([df_class_1_over, df_class_0], axis = 0)

print("Random over-sampling:")
print(df_test_over.Promoted_or_Not.value_counts())
df_test_over.Promoted_or_Not.value_counts().plot(kind = "bar", title = "Count(target)")
```
